```python
from dataclasses import field
from urllib import response
from bs4 import BeautifulSoup
import requests
import re
import csv
import lxml

url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
print(response)

movies = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
votes = [b.attrs.get('data-value') for b in soup.select('td.ratingColumn strong')]

list = []

# Iterating over movies to extract
# each movie's details
for index in range(0, len(movies)):
    # Separating  movie into: 'place',
    # 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index)) + 1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index)) - (len(movie))]

    data = {"movie_title": movie_title,
            "year": year,
            "place": place,
            "star_cast": crew[index],
            "rating": ratings[index],
            "vote": votes[index],
            "link": links[index]}
    list.append(data)

for movie in list:
    # print(movie['place'], '-', movie['movie_title'], '('+movie['year'] +
    #       ') -', 'Starring:', movie['star_cast'], movie['rating'])

    field_names = ['movie_title', "year", "place", "star_cast", "rating", "vote", "link"]
    with open('movies.csv', 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=field_names)
        writer.writeheader()
        writer.writerows(list)

print("successful")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract data for the top 10 movies
top_10_movies = list[:10]
top_10_titles = [movie['movie_title'] for movie in top_10_movies]
top_10_ratings = [float(movie['rating']) for movie in top_10_movies]

# Set color scheme
colors = plt.cm.get_cmap('tab20c')

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(12, 6))

# Plot horizontal bar chart
bars = ax.barh(np.arange(len(top_10_titles)), top_10_ratings, color=colors(range(len(top_10_ratings))))

# Add labels and annotations
ax.set_yticks(np.arange(len(top_10_titles)))
ax.set_yticklabels(top_10_titles, fontsize=12)
ax.set_xlabel('Rating', fontsize=12)
ax.set_title('Top 10 Movie Ratings', fontsize=14, fontweight='bold')

# Add value annotations to the bars
for i, bar in enumerate(bars):
    rating = top_10_ratings[i]
    ax.text(rating + 0.1, bar.get_y() + bar.get_height() / 2, f'{rating:.2f}', ha='left', va='center', fontsize=11)

# Customize the appearance
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(axis='both', which='both', length=0)
ax.set_xlim([0, max(top_10_ratings) + 1])
ax.xaxis.grid(True, linestyle='--', color='lightgray')

# Set color for y-axis labels
ax.set_yticklabels(top_10_titles, fontsize=12, color='gray')

# Set chart background color
ax.set_facecolor('#f9f9f9')

# Set bar colors
for bar in bars:
    bar.set_edgecolor('white')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Create DataFrame from the movie data
df = pd.DataFrame(list)

# Convert 'rating' and 'vote' columns to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['vote'] = pd.to_numeric(df['vote'], errors='coerce')

# Sort DataFrame by rating in descending order
df = df.sort_values(by='rating', ascending=False)

# Set up the figure and axes
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Top Rated Movies', fontsize=16, fontweight='bold')


# Chart 1: Histogram - Distribution of Ratings
axs[0, 0].hist(df['rating'], bins=20, color='lightgreen', edgecolor='black')
axs[0, 0].set_title('Distribution of Ratings', fontsize=12, fontweight='bold')
axs[0, 0].set_xlabel('Rating')
axs[0, 0].set_ylabel('Frequency')

# Chart 2: Box Plot - Rating Distribution
axs[0, 1].boxplot(df['rating'], patch_artist=True, boxprops=dict(facecolor='lightgray'))
axs[0, 1].set_title('Rating Distribution', fontsize=12, fontweight='bold')
axs[0, 1].set_ylabel('Rating')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```
